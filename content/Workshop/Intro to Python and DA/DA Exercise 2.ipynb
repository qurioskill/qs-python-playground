```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Calculate unit price
df['UNITPRICE'] = df['SALES'] / df['QUANTITYORDERED']

# Drop invalid rows
df = df[(df['UNITPRICE'] > 0) & (df['QUANTITYORDERED'] > 0)]

# Prepare variables
x = df['UNITPRICE']
y = df['QUANTITYORDERED']

# Simple linear regression using numpy
slope, intercept = np.polyfit(x, y, 1)

# Predict y values
y_pred = slope * x + intercept

# Plot
plt.figure(figsize=(8, 5))
plt.scatter(x, y, alpha=0.4, label='Data')
plt.plot(x, y_pred, color='red', label=f'Fit: y={slope:.2f}x + {intercept:.2f}')
plt.title('Unit Price vs Quantity Ordered')
plt.xlabel('Unit Price')
plt.ylabel('Quantity Ordered')
plt.legend()
plt.grid(True)
plt.show()

# Display correlation coefficient
corr = np.corrcoef(x, y)[0, 1]
print(f"Correlation between Unit Price and Quantity Ordered: {corr:.2f}")
```
